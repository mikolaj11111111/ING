```python
import pandas as pd
import folium
import geopandas as gpd
import json
import os
```

```python
pd.options.display.float_format = '{:.2f}'.format
quarters = []
```

```python
quarter1 = pd.read_excel('Dzielnice Warszawy.xlsx', sheet_name="Sheet2")[:1]
quarter1.drop(columns=["Date", "Actual"], inplace=True)
quarters.append(quarter1)

quarter2 = pd.read_excel('Dzielnice Warszawy.xlsx', sheet_name="Sheet2")[1:2]
quarter2.drop(columns=["Date", "Actual"], inplace=True)
quarters.append(quarter2)

quarter3 = pd.read_excel('Dzielnice Warszawy.xlsx', sheet_name="Sheet2")[2:3]
quarter3.drop(columns=["Date", "Actual"], inplace=True)
quarters.append(quarter3)

quarter4 = pd.read_excel('Dzielnice Warszawy.xlsx', sheet_name="Sheet2")[3:]
quarter4.drop(columns=["Date", "Actual"], inplace=True)
quarters.append(quarter4)

history_data = pd.read_excel('Dzielnice Warszawy.xlsx', sheet_name="Sheet1")[:20:]

mean_data = pd.read_excel('Dzielnice Warszawy.xlsx', sheet_name="Sheet1").iloc[[29], :]
mean_data.drop(columns=["Date"], inplace=True)

metric_data = pd.read_excel('Dzielnice Warszawy.xlsx', sheet_name="Sheet1").iloc[[30], :]
metric_data.drop(columns=["Date", "Actual"], inplace=True)
```

```python

coords = {
    "Bemowo": (52.254421, 20.931117),
    "Białołęka": (52.335, 21.034),
    "Bielany": (52.282511, 20.958614),
    "Mokotów": (52.189577, 21.052186),
    "Ochota": (52.208024, 20.992977),
    "Praga-Południe": (52.238551, 21.107485),
    "Praga-Północ": (52.258861, 21.066019),
    "Rembertów": (52.257838, 21.177325),
    "Targówek": (52.271752, 21.089594),
    "Ursus": (52.194617, 20.911365),
    "Ursynów": (52.128, 21.060),
    "Wawer": (52.191265, 21.179828),
    "Wesoła": (52.226026, 21.260604),
    "Wilanów": (52.141161, 21.127698),
    "Wola": (52.238408, 20.991004),
    "Włochy": (52.183836, 20.982670),
    "Śródmieście": (52.232230, 21.052091),
    "Żoliborz": (52.264265, 21.015249)
}
```

```python
districts = [
    "bemowo", "bialoleka", "bielany", "mokotow", "ochota", 
    "praga-poludnie", "praga-polnoc", "rembertow", "targowek", 
    "ursus", "ursynow", "wawer", "wesola", "wilanow", 
    "wola", "wlochy", "srodmiescie", "zoliborz"
]
```

```python
m = folium.Map(location=[52.2298, 21.0122], zoom_start=11)

def put_price_on_map(quarter):
    
    for area, cord in coords.items():
        folium.map.Marker(
            cord,
            icon=folium.DivIcon(
                icon_size=(150, 36),
                icon_anchor=(75, 18),
                html=f'<div style="font-size: 12pt; color: black; font-weight: bold;">{int(quarter[area].values[0])}</div>',
            ),
        ).add_to(m)
    for district in districts:
        geojson_path = f"{district}.geojson"

        with open(geojson_path, "r", encoding="utf-8") as f:
            geojson_loc = json.load(f)
    
            folium.GeoJson(
                geojson_loc,
                name="{district}",
                style_function=lambda feature: {
                    "fillColor": "white",
                    "color": "black",
                    "weight": 2,
                    "fillOpacity": 0.4
                }
            ).add_to(m)
    return m
```

```python
maps = [put_price_on_map(quarter) for quarter in quarters]
```
